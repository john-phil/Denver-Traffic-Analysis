```python
# Convert the 'ROAD_CONDITION' column to lowercase to normalize the data
df_cleaned['ROAD_CONDITION'] = df_cleaned['ROAD_CONDITION'].str.lower()

# Count the frequency of each unique road condition
road_conditions_counts = df_cleaned['ROAD_CONDITION'].value_counts()

# Calculate the total number of incidents
total_incidents = df_cleaned.shape[0]

# Convert counts to percentages
road_conditions_percentages = (road_conditions_counts / total_incidents) * 100

# Display the percentage of each road condition
road_conditions_percentages
```

```python
# Filter accidents where there were seriously injured or fatalities
serious_or_fatal = df_cleaned[(df_cleaned['SERIOUSLY_INJURED'] > 0) | (df_cleaned['FATALITIES'] > 0)]

# Count the frequency of each unique road condition in serious or fatal accidents
serious_or_fatal_counts = serious_or_fatal['ROAD_CONDITION'].value_counts()

# Calculate the total number of serious or fatal incidents
total_serious_or_fatal = serious_or_fatal.shape[0]

# Convert counts to percentages
serious_or_fatal_percentages = (serious_or_fatal_counts / total_serious_or_fatal) * 100

# Get overall road conditions percentages from previous analysis
# Assuming 'road_conditions_percentages' is already defined
# Here we also calculate percentages for all incidents to compare
total_road_conditions_counts = df_cleaned['ROAD_CONDITION'].value_counts()
total_incidents = df_cleaned.shape[0]
overall_road_conditions_percentages = (total_road_conditions_counts / total_incidents) * 100

# Create a DataFrame for comparison
comparison_df = pd.DataFrame({
    'Overall_Percent': overall_road_conditions_percentages,
    'Serious_or_Fatal_Percent': serious_or_fatal_percentages
})

# Calculate the relative risk ratio
comparison_df['Relative_Risk'] = comparison_df['Serious_or_Fatal_Percent'] / comparison_df['Overall_Percent']

# Display the result
comparison_df_sorted=comparison_df.sort_values(by='Relative_Risk', ascending=False)

comparison_df_sorted.to_clipboard()

comparison_df_sorted
```

```python
# Calculate the frequency of each traffic offense
offense_counts = df_cleaned['top_traffic_accident_offense'].value_counts()

# Calculate the total number of incidents
total_offenses = offense_counts.sum()

# Calculate the percentage of each offense
offense_percentages = (offense_counts / total_offenses) * 100
```

```python
# Filter accidents where there were seriously injured or fatalities
serious_or_fatal_accidents = df_cleaned[(df_cleaned['SERIOUSLY_INJURED'] > 0) | (df_cleaned['FATALITIES'] > 0)]
```

```python
# Count the frequency of each offense in serious or fatal accidents
serious_or_fatal_offense_counts = serious_or_fatal_accidents['top_traffic_accident_offense'].value_counts()

# Calculate the total number of serious or fatal incidents
total_serious_or_fatal = serious_or_fatal_offense_counts.sum()

# Convert counts to percentages
serious_or_fatal_offense_percentages = (serious_or_fatal_offense_counts / total_serious_or_fatal) * 100
```

```python
# Create a DataFrame for comparison
offense_comparison_df = pd.DataFrame({
    'Overall_Percent': offense_percentages,
    'Serious_or_Fatal_Percent': serious_or_fatal_offense_percentages
})

# Calculate the relative risk ratio
offense_comparison_df['Relative_Risk'] = offense_comparison_df['Serious_or_Fatal_Percent'] / offense_comparison_df['Overall_Percent']

# Display the result
offense_comparison_df_sorted=offense_comparison_df.sort_values(by='Relative_Risk', ascending=False)

offense_comparison_df_sorted.to_clipboard()

offense_comparison_df_sorted
```

```python
# Calculate the frequency of incidents in each neighborhood
neighborhood_counts = df_cleaned['neighborhood_id'].value_counts()

# Calculate the total number of incidents
total_neighborhood_incidents = neighborhood_counts.sum()

# Calculate the percentage of each neighborhood
neighborhood_percentages = (neighborhood_counts / total_neighborhood_incidents) * 100
```

```python
# Filter accidents where there were seriously injured or fatalities
serious_or_fatal_accidents = df_cleaned[(df_cleaned['SERIOUSLY_INJURED'] > 0) | (df_cleaned['FATALITIES'] > 0)]
```

```python
# Count the frequency of each neighborhood in serious or fatal accidents
serious_or_fatal_neighborhood_counts = serious_or_fatal_accidents['neighborhood_id'].value_counts()

# Calculate the total number of serious or fatal incidents
total_serious_or_fatal_neighborhood = serious_or_fatal_neighborhood_counts.sum()

# Convert counts to percentages
serious_or_fatal_neighborhood_percentages = (serious_or_fatal_neighborhood_counts / total_serious_or_fatal_neighborhood) * 100
```

```python
# Set the option to display more rows, you can adjust the number as needed
pd.set_option('display.max_rows', None) 

# Create a DataFrame for comparison
neighborhood_comparison_df = pd.DataFrame({
    'Overall_Percent': neighborhood_percentages,
    'Serious_or_Fatal_Percent': serious_or_fatal_neighborhood_percentages
})

# Calculate the relative risk ratio
neighborhood_comparison_df['Relative_Risk'] = neighborhood_comparison_df['Serious_or_Fatal_Percent'] / neighborhood_comparison_df['Overall_Percent']

# Display the result, sorting by the relative risk to see the highest risk neighborhoods
neighborhood_comparison_df_sorted=neighborhood_comparison_df.sort_values(by='Relative_Risk', ascending=False)

neighborhood_comparison_df_sorted.to_clipboard()

neighborhood_comparison_df_sorted
```
